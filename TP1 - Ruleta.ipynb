### Importacion de paquetes

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Experimento: 100 tiradas

#### Generacion experimento

```python
res = np.random.randint(low= 0, high= 37, size= 100)
med_gen = np.mean(res)
print(med_gen)
std_gen = np.std(res)
print(std_gen)
```

#### Frecuencias relativas

```python
plt.hist(res, bins= 37, density = True, color= 'red')
plt.title("Distribucion de las frecuencias relativas")
plt.xlabel("Valor")
plt.ylabel("Frecuencia relativa")
plt.show()
```

#### Esperanza

```python
prom = [res[0], res[1]]
media = [np.median(prom)]
for idx, val in enumerate(res):
    if idx > 1:
        prom.append(val)
        med = np.median(prom)
        media.append(med)
plt.plot(media)
plt.show()
```

#### Desvio

```python
prom = [res[0], res[1]]
desv = [np.std(prom)]
for idx, val in enumerate(res):
    if idx > 1:
        prom.append(val)
        des = np.std(prom)
        desv.append(des)
plt.plot(desv)
plt.show()
```
