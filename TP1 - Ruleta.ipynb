## Importacion de paquetes

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Promedio

```python
si = 10000        # Tamao de la muestra
ex = 5             # cantidad de experimentos
vi = 0              # valor inicial ruleta
vf = 36             # valor final ruleta
et = (vf - vi - 1) / 2  # Promedio teorico
med_et = []         # arreglo del promedio teorico

for i in range(ex):
    res = np.random.randint(low= vi, high= vf, size= si)
    prom = []
    media = []
    # print(res)

    for idx, val in enumerate(res):
        
        prom.append(val)
        med = np.mean(prom)
        media.append(med)
    plt.plot(media)

for i in range(0, si):
    med_et.append(et)

plt.plot(med_et, color= "black", label= "Valor teorico")
plt.axis([0, si, et - 5, et + 5])
plt.title("Distribucion de los promedios")
plt.xlabel("NRO de tirada")
plt.ylabel("Promedio")
plt.legend(loc="upper right")
plt.show()
```

## Frecuencia Relativa

```python
si = 1000  # Tamao de la muestra
ex = 10 # Cantidad de experimentos
vi = 0 # Valor inicial
vf = 36 # Valor Final
x = 7               # nmero del experimento
vt = 1/(vf - vi + 1)  # valor terico
array_fr = []       # frecuencia relativa
array_vt = []
counti = 0
countx = 0
maximo = 0

for x in range(si):
    counti += 1
    if np.random.randint(vi, vf) == x:
        countx += 1
    if (countx/counti)>maximo:
        maximo=countx/counti
    array_fr.append(countx/counti)
    array_vt.append(vt)

plt.title('Frecuencia relativa del valor '+ repr(x))
plt.plot(array_vt, label = "Valor Teorico")
plt.plot(array_fr, label = "Valor muestral")
plt.xlabel('Tirada')
plt.ylabel('Frecuencia Relativa')
plt.axis([0, si, -0.01, maximo + 0.01])
plt.legend(loc="upper left")
plt.show()
    
```
