## Importacion de paquetes

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Promedio

```python
si = 500                # Tamaño de la muestra
ex = 5                  # cantidad de experimentos
vi = 0                  # valor inicial ruleta
vf = 36                 # valor final ruleta
et = (vf - vi - 1) / 2  # Promedio teorico
med_et = []             # arreglo del promedio teorico

for i in range(ex):
    res = np.random.randint(low= vi, high= vf, size= si)
    prom = []
    media = []

    for idx, val in enumerate(res):
        
        prom.append(val)
        med = np.mean(prom)
        media.append(med)
    plt.plot(media)

for i in range(0, si):
    med_et.append(et)

plt.plot(med_et, color= "black", label= "Valor teorico")
plt.axis([0, si, et - 5, et + 5])
plt.title("Distribucion de los promedios")
plt.xlabel("NRO de tirada")
plt.ylabel("Promedio")
plt.legend(loc="upper right")
plt.show()
```

```python
tm = 5000               # Tamaño de la muestra
ex = 5                  # cantidad de experimentos
vi = 0                  # valor inicial ruleta
vf = 36                 # valor final ruleta
et = (vf - vi ) / 2     # Esperanza teorica
med_et = []             # arreglo del promedio teorico

for i in range(ex):    # Bucle para cada experimento

    arr_nro = []
    arr_prom = []
    for tir in range(1, tm + 1):    # Bucle para cada tirada
        nro_random = np.random.randint(vi, vf + 1)
        arr_nro.append(nro_random)
        prom = np.mean(arr_nro)
        arr_prom.append(prom)
    plt.plot(arr_prom)    # Grafico del experimento

for tir in range(1, tm +1): # Bucle adicional para graficar el valor teorico
    med_et.append(et)

# a continuacion grafico del valor teorico y detalles adicionales de graficacion
plt.plot(med_et, color= "black", label= "Valor teorico")
plt.axis([0, tm, et - 5, et + 5])
plt.title("Distribucion de los promedios")
plt.xlabel("NRO de tirada")
plt.ylabel("Promedio")
plt.legend(loc="upper right")
plt.show()
```

## Frecuencia Relativa

```python
tm = 5000              # Tamaño de la muestra
ex = 5                # cantidad de experimentos
vi = 0                  # valor inicial ruleta
vf = 36                 # valor final ruleta
x = 5                   # Numero a evaluar
ft = 1/(vf - vi + 1)    # Frecuencia teorica
arr_ft = []

for i in range(ex):    # Bucle para cada experimento
    countx = 0         # contador de cuantas veces salio x
    countt = 0         # contador de totales
    arr_frec = []      # arreglo de frecuencias

    for tir in range(1, tm + 1):    # Bucle para cada tirada
        nro_random = np.random.randint(vi, vf + 1)
        countt += 1

        if nro_random == x:
            countx += 1
        arr_frec.append(countx/countt)
    
    plt.plot(arr_frec)

for tir in range(1, tm + 1):    # Bucle para cada tirada
    arr_ft.append(ft)

plt.plot(arr_ft, color= "black", label= "Frecuencia teorica")
plt.title("Frecuencia relativa de un numero X")
plt.xlabel("NRO de tirada")
plt.ylabel("Frecuencia relativa")
plt.legend(loc="upper right")
plt.axis([0, tm, ft - 0.04, ft + 0.04])
plt.show()
```
