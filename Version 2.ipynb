## Histograma

```python
si = 100  # Tamaño de la muestra
ex = 2   # Cantidad de experimentos

for i in range(0, ex):  
    res = np.random.randint(low= 0, high= 37, size= si)
    plt.hist(res, bins= 37, density= True)

plt.title("Distribucion de las Frecuencias Relativas")
plt.xlabel("Valor")
plt.ylabel("Frecuencia relativa")
plt.show()
```
