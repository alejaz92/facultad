## Importamos librerias

```python
import numpy as np 
import matplotlib.pyplot as plt
```

## Esperanza

```python
si = 1000  # Tamao de la muestra
ex = 50   # Cantidad de experimentos

for i in range(0, ex):
    res = np.random.randint(low= 0, high= 37, size= si)
    prom = [res[0], res[1]]
    media = [np.median(prom)]
    for idx, val in enumerate(res):
        if idx > 1:
            prom.append(val)
            med = np.median(prom)
            media.append(med)
    plt.plot(media)
plt.title("Distribucion de los promedios")
plt.xlabel("Valor")
plt.ylabel("Frecuencia relativa")
plt.show()
```
