## Importamos librerias

```python
import numpy as np 
import matplotlib.pyplot as plt
```

## Esperanza

```python
si = 1000  # Tamao de la muestra
ex = 1  # Cantidad de experimentos
vi = 0 # Valor inicial
vf = 36 # Valor Final
et = (vf - vi)/2
med_et = []

for i in range(0, ex):
    res = np.random.randint(low= vi, high= vf, size= si)
    prom = [res[0], res[1]]
    media = [np.median(prom)]

    for idx, val in enumerate(res):
        if idx > 1:
            prom.append(val)
            med = np.median(prom)
            media.append(med)
    plt.plot(media)

for idx, val in enumerate(res):
    med_et.append(et)

plt.plot(med_et, color= 'black')
plt.title("Distribucion de los promedios")
plt.xlabel("Nro Tirada")
plt.ylabel("Promedio")
plt.show()
```

## Frecuencia Relativa

```python
si = 100  # Tamao de la muestra
ex = 1   # Cantidad de experimentos
x = 0    # Valor x

for i in range(0, ex):
    res = np.random.randint(low= 0, high= 37, size= si)
    conta_x = 0
    for idx, val in enumerate(res):
        if val == x:
            conta_x += 1
            plt.plot(conta_x)
            plt.show()
```

## Histograma

```python
si = 100  # Tamao de la muestra
ex = 2   # Cantidad de experimentos

for i in range(0, ex):  
    res = np.random.randint(low= 0, high= 37, size= si)
    plt.hist(res, bins= 37, density= True)

plt.title("Distribucion de las Frecuencias Relativas")
plt.xlabel("Valor")
plt.ylabel("Frecuencia relativa")
plt.show()
```

## Desvio Estndar

```python
si = 5000  # Tamao de la muestra
ex = 10  # Cantidad de experimentos

for i in range(0, ex):
    res = np.random.randint(low= 0, high= 37, size= si)
    prom = [res[0], res[1]]
    desv = [np.std(prom)]
    for idx, val in enumerate(res):
        if idx > 1:
            prom.append(val)
            des = np.std(prom)
            desv.append(des)
    plt.plot(desv)

plt.title("Distribucion de los Desvios")
plt.xlabel("Nro Tirada")
plt.ylabel("Desvio Estndar")
plt.show()
```
