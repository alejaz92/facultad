Generador Numpy

```python
import numpy as np 

def np_method(m):
    num_list = []
    for i in range(m):
        res = np.random.randint(1, 10000)
        num_list.append(round(res/10000, 4))
    return(num_list)
```

# Generador Normal

```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
import scipy.stats as stats

i = 10000
a = 0
b = 10

e = math.e
pi = math.pi
data = []

EX = (a+b)/2

for x in range(i):
    sum = 0

    for y in range(12):
        r = np.random.uniform(a, b)
        sum = sum + r
    X = 0.1*(sum-6.0) + EX
    data.append(X)

python = np.random.normal(10, 1, i)
bins = 10

mu = 10
variance = 2
sigma = math.sqrt(variance)
x = np.linspace(mu - 6*sigma, mu + 6*sigma, 100)

plt.hist(data, bins, edgecolor='black', weights=np.zeros_like(data) + 1./len(data), alpha=0.5, label="Generador de Python")
plt.hist(python, bins, edgecolor='black', weights=np.zeros_like(python) + 1./len(python), alpha=0.5, label="Generador de Python")
plt.plot(x, stats.norm.pdf(x, mu, sigma), label="Normal Teórica")
plt.legend(loc="upper right")

plt.show()
```

## Distribucion Uniforme

### Importaciones y parametros

```python
import numpy as np 
import matplotlib.pyplot as plt

m = 100000 # tamaño de la muestra
a = 0  # limite inferior
b = 10  # limite superior
bins = 100
```

### Transformada inversa

```python
r_array = np_method(m)

x_arr = []

for i in range(m):
    x = a + (b - a) * r_array[i]
    x_arr.append(x)
plt.hist(x_arr, bins = bins, edgecolor= "black")
plt.title("Distribucion Uniforme - Transformada Inversa")
plt.axis([a, b, 0, 1200])
plt.xlabel("Rango")
plt.ylabel("Frecuencia")
plt.show()
```

### Metodo Numpy

```python
uni_np_arr = []
for i in range(m):
    uni_np_arr.append(np.random.uniform(a, b))
plt.hist(uni_np_arr, bins = bins, color= "green", edgecolor= "black")
plt.title("Distribucion Uniforme - Metodo Numpy")
plt.axis([a, b, 0, 1200])
plt.xlabel("Rango")
plt.ylabel("Frecuencia")
plt.show()
```

### Valor teorico

```python
plt.plot([a, b], [m/bins, m/bins], color='black', label='valor teorico', linewidth=4)
plt.title("Distribucion Uniforme - Valor Teorico")
plt.axis([a, b, 0, 1200])
plt.xlabel("Rango")
plt.ylabel("Frecuencia")
plt.show()
```

### Comparativa de metodos

```python
plt.hist(x_arr, bins = bins, edgecolor= "black", label= "Transformada Inversa")       # transformada inversa
plt.hist(uni_np_arr, bins = bins, color= "green", edgecolor= "black", label= "Metodo Numpy")                 #  metodo numpy
plt.plot([a, b], [m/bins, m/bins], color='black', label='valor teorico', linewidth=4) # teorico

plt.axis([a, b, 0, 1200])
plt.title("Distribucion Uniforme - Comparativa de metodos")
plt.xlabel("Rango")
plt.ylabel("Frecuencia")
plt.legend(loc="lower center")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

i = 2000

a=0
b=10
data = []
python = []

for x in range(i):
    z = np.random.uniform(a,b)
    r = np.random.uniform(0, 1)
    y = 0
    y = a + (b-a)*r
    data.append(y)
    python.append(z)
    #print(r, y)

bins = 20
plt.hist(python, bins, edgecolor='black', alpha=0.5, label="Generador de Python")
plt.hist(data, bins, edgecolor='black', alpha=0.5, label="Algoritmo Generador")

plt.plot([0, 10], [i/20, i/20], color='black', label='Uniforme teorica (0,10)', linewidth=4)
plt.legend()
plt.legend(loc="lower center")
plt.show()
```

## Distribucion Exponencial

```python
import numpy as np 
import matplotlib.pyplot as plt 
from math import log

m = 1000000     # nro de muestras

alpha = 0.05      

r_array = np_method(m)
#print(r_array)
x_arr = []
x = 0
for i in range(m):
    r = r_array[i]
    x = (- log(1 - r)) / alpha
    x_arr.append(x)
plt.hist(x_arr, bins= 100)
plt.title("Distribucion Exponencial")
plt.xlabel("Rango")
plt.ylabel("Frecuencia")
plt.show()
```
